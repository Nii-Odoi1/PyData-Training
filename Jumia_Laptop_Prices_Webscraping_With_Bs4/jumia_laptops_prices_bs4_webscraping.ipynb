```python
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
url = 'https://www.jumia.com.gh/laptops/'
jumia_request = Request(url, headers={'User-Agent': 'Chrome/5.0'})
page = urlopen(jumia_request).read()
page
```

```python
#create an instance of the BeautifulSoup class to parse our document
jumia_soup = soup(page, 'html.parser')

#Print out the HTML content of the page, formatted nicely, using the prettify method on the BeautifulSoup object
print(jumia_soup.prettify())
```

```python
products = [pro.get_text() for pro in jumia_soup.select(".info .name")]
products
```

```python
price = [cost.get_text() for cost in jumia_soup.select(".info .prc")]
price
```

```python
len(price)
```

```python
len(products)
```

```python
df = pd.DataFrame({
    "Laptop Name": products,
    "Price": price
})
df
```

```python
#to keep only the rows with data
final_output = df[:40]
final_output
```

```python
final_output.to_csv('Jumia_Laptops_Prices.csv')
```
