```python
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
url = 'https://www.jumia.com.gh/phones-tablets/'
jumia_request = Request(url, headers={'User-Agent': 'Chrome/5.0'})
page = urlopen(jumia_request).read()
page
```

```python
# Using the BeautifulSoup library to parse this document
#creating an instance of the BeautifulSoup class to parse the document
jumia_phones_soup = soup(page, 'html.parser')
#Print out the HTML content of the page, formatted nicely, using the prettify method on the BeautifulSoup object
print(jumia_phones_soup.prettify())
```

```python
products = [phone.get_text() for phone in jumia_phones_soup.select(".core .name")]
products
```

```python
price = [cost.get_text() for cost in jumia_phones_soup.select(".core .prc")]
price
```

```python
len(price)
```

```python
len(products)
```

```python
df = pd.DataFrame({
    "Phones & Accessories": products,
    "Price": price
})
df
```

```python
phone_final_output = df[:145]
phone_final_output
```
