```python
june_16['date'] = june_16['date'].dt.strftime('%B %d, %Y')
june_16
```

```python
rollig_dataset = df.iloc[485:492]
rollig_dataset
```

```python
new_rolling_avg = rollig_dataset['actuals.newCases'].mean()
actual_rolling_avg = rollig_dataset['actuals.cases'].mean()
cases_100k = (new_rolling_avg/actual_rolling_avg)*100000
rounded_rolling_avg = round(cases_100k, 2)
rounded_rolling_avg
```

```python
actual_cases = int(june_16['actuals.cases'])
actual_deaths = int(june_16['actuals.deaths'])
state = june_16['state'].to_string(index=False)
june_16_date = june_16['date'].astype('str').to_string(index=False)
june_16_date_up = june_16_date.replace(',','th')
new_cases = int(june_16['actuals.newCases'])
new_deaths = int(june_16['actuals.newDeaths'])
cummulative_cases = actual_cases + new_cases
cummulative_deaths = actual_deaths + new_deaths
cases_per_100k = rounded_rolling_avg
vacc_comp_ratio = float(june_16['metrics.vaccinationsCompletedRatio'])
vacc_comp_fig = int(june_16['actuals.vaccinationsCompleted'])
vacc_init_ratio = float(june_16['metrics.vaccinationsInitiatedRatio'])
vacc_init_fig = int(june_16['actuals.vaccinationsInitiated'])
```

```python
display = f"""##############################################################

{state}, {june_16_date_up}

Infection
        - New cases: {new_cases}
        - New deaths: {new_deaths}
        - Cummulative cases: {cummulative_cases}
        - Cummulative deaths: {cummulative_deaths}
        - Cases per 100k (using a 7-day rolling average): {cases_per_100k}
        
        Vaccination
        - Completed: {vacc_comp_ratio} ({vacc_comp_fig})
        - Initiated: {vacc_init_ratio} ({vacc_init_fig})

##############################################################"""
print (display)
```

# Relationship Between Actual COVID Cases & Vaccination Completed

1. There's a positive relationship between the actual COVID cases and the vaccination completed. i.e as the covid  cases increase, vaccination completed also increase and when the cases decrease, vaccination completed also decrease.

2. From 2nd to 5th March 2020, when there were no cases, vaccination completed dropped completely to 0 and started increasing when cases started rising again.
